```python
import pints
import pints.toy as toy

import emupints
import emupints.plot as emuplt
import emupints.utils as emutils

import matplotlib.pyplot as plt
import numpy as np

import string
```

```python
# Load problem from predefined ones
model = emupints.Problems.LogisticModel
problem = emupints.Problems.load_problem(model)
```

```python
# take required variables for visualisation
model = problem['model']
values = problem['values']
times = problem['times']
parameters = problem['parameters']
```

```python
synth_data = model.simulate(parameters, times)
```

```python
help(model.simulate)
```

```python
# Customise the plot, and add the original, noise-free data
plt.figure(figsize=(10, 4))
plt.xlabel('Time')
plt.ylabel('Population Size')
plt.plot(
    times[::2], values[::2], 'x', color='black', ms=6, alpha=0.75,
    label='Observed data')
plt.plot(
    times[::2], synth_data[::2], color='#1f77b4',
    label='Underlying model')
plt.legend()
```
