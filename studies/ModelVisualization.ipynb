This notebook can be used to visualise the surfaces of the model

```python
import pints
import pints.toy as toy

import emupints
import emupints.plot as emuplt
import emupints.utils as emutils

import matplotlib.pyplot as plt
import numpy as np

import string
```

```python
# Load problem from predefined ones
model = emupints.Problems.SIRModel
problem = emupints.Problems.load_problem(model)

problem['values'][:5]
```

```python
# take required variables for visualisation
n_parameters = problem['n_parameters']
log_likelihood = problem['log_likelihood']
bounds = problem['bounds']
index_to_param_name = problem['param_names']
```

```python
log_likelihood = pints.UnknownNoiseLogLikelihood(problem['problem'])
```

```python
help(log_likelihood([0.15, 3, 2]))
```

## Visualisations

```python
fig, ax = emuplt.plot_surface_fixed_param(
    log_likelihood,
    bounds,
#    fixed=[(0, 0.10)],
    contour=False,
    index_to_param_name=index_to_param_name,
    precision=2,
)
```

```python
ax.view_init(30, 45 + 75)
fig
```

```python
fixed_parameters = emutils.fix_parameters(bounds)
```

```python
if n_parameters > 2:  
    fig, ax = emuplt.plot_fixed_param_grid(
        log_likelihood,
        fixed_parameters,
        bounds,
        n_splits = 50, # number of splits on each free parameter
        shape = (n_parameters, n_parameters - 1),
        index_to_param_name=index_to_param_name
    )

    plt.show(fig)
```

```python
if n_parameters > 2:  
    fig, ax = emuplt.plot_fixed_param_grid(
        log_likelihood,
        fixed_parameters,
        bounds,
        n_splits = 50, # number of splits on each free parameter
        shape = (n_parameters, n_parameters - 1),
        index_to_param_name=index_to_param_name,
        contour=False
    )

    plt.show(fig)
```
